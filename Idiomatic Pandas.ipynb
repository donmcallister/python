```python
# a glorious function
def tweak_autos(autos):
    return (autos
     [cols]
     .assign(
         cylinders=autos.cylinders.fillna(0).astype('int8'), #overwrite or add new column!
         displ    =autos.displ.fillna(0),
         drive    =autos.drive.fillna("Other").astype('category'),
         automatic=autos.trany.str.contains('Auto'),
         speeds   =autos.trany.str.extract(r'(\d)+').fillna('20').astype('int8'),
         createdOn=pd.to_datetime(autos.createdOn.replace({' EDT':"-04:00",
                                                          ' EST': "-05:00"}, regex=True)),
         ffs      =autos.eng_dscr.str.contains('FFS')
        )
     .astype({'city08':'int16', 'comb08': 'int16', 'highway08':'int8',
          'fuelCostA08':'int16', 'year': 'int16', 'make':'category'})
     .drop(columns=['trany', 'eng_dscr'])
    )

tweak_autos(autos).head(5)
```

# Chain and Pipe

```python
def get_var(df, var_name):
    globals()[var_name] = df
    return df

def tweak_autos(autos):
    return (autos
     [cols]
      #create var
     .pipe(get_var, 'df3')
     .assign(
         cylinders=autos.cylinders.fillna(0).astype('int8'), #overwrite or add new column!
         displ    =autos.displ.fillna(0),
         drive    =autos.drive.fillna("Other").astype('category'),
         automatic=autos.trany.str.contains('Auto'),
         speeds   =autos.trany.str.extract(r'(\d)+').fillna('20').astype('int8'),
         createdOn=pd.to_datetime(autos.createdOn.replace({' EDT':"-04:00",
                                                          ' EST': "-05:00"}, regex=True)),
         ffs      =autos.eng_dscr.str.contains('FFS')
        )
     #debug pipe
     .pipe(lambda df: display(df) or df)
     .astype({'city08':'int16', 'comb08': 'int16', 'highway08':'int8',
          'fuelCostA08':'int16', 'year': 'int16', 'make':'category'})
     .drop(columns=['trany', 'eng_dscr'])
    )

tweak_autos(autos).head(5)
```

```python
def country(val):
    if val in {'Chevrolet', 'Ford', 'Dodge', 'GMC', 'Tesla'}:
        return 'US'
    return 'Other'
```

```python
(autos2
 .assign(country=autos2.make.apply(country))
 .groupby(['year', 'country'])
 .mean()
).sample(5)
```

```python
def second_to_last(ser):
    return ser.iloc[-2]

(autos2
 .assign(country=autos2.make.apply(country))
 .groupby(['year', 'country'])
 .agg(['min', 'mean', second_to_last])
).sample(5)
```

```python
(autos2
 .assign(country=autos2.make.apply(country))
 .groupby(['year', 'country'])
 .mean()
 .unstack()
 .city08
 .rolling(2)
 .mean()
).sample(5)
```
