# Python Pandas Expand a Column of List of Lists to Two New Column
- https://stackoverflow.com/questions/56095142/python-pandas-expand-a-column-of-list-of-lists-to-two-new-column

```python
import pandas as pd
df = pd.DataFrame({
    'name': ['john', 'smith'],
    'id': [1, 2],
    'apps': [[['app1', 'v1'], ['app2', 'v2'], ['app3','v3']], 
             [['app1', 'v1'], ['app4', 'v4']]]
})

print("df.apps:\n", df.apps)

dftmp = df.apps.apply(pd.Series)
print("\n==> apply(pd.Series):\n", dftmp)

dftmp = dftmp.T.melt().dropna()
print("\n==> .T.melt().dropna():\n", dftmp)
print("\n==> .value:\n", dftmp.value)
print("\n==> .variable:\n", dftmp.variable)

dftmp = dftmp.apply(pd.Series)
print("\n==> dftmp.apply(pd.Series):\n", dftmp)

dfapp = (dftmp.value
              .apply(pd.Series)
              .set_index(dftmp.variable)
              .rename(columns={0:'app_name', 1:'app_version'})
        )
print("\n==> dfapp dataframe from dftmp.value:\n",dfapp)

df[['name', 'id']].merge(dfapp, left_index=True, right_index=True)
```
